```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_csv("C:/Users/USER/DSI-course-materials/curriculum/04-lessons/week-01/3.2-lesson/assets/datasets/sales.csv")
```

```python
sales_data.head()
sales_data["Total Revenue"] = sales_data["Quantity"]*sales_data["Price"]
```

```python
#creates list of reps for IDs
rep_list = sales_data["Rep"].unique().tolist()
print(rep_list)
```

```python
#Adds Rep ID into data frame
print(rep_list.index(sales_data["Rep"][10]))
rep_id_list = []
for i in sales_data["Rep"]:
    rep_id_list.append(rep_list.index(i))
print(rep_id_list)
sales_data["Rep ID"] = rep_id_list
```

```python
sales_data_pivot = pd.pivot_table(sales_data, index=['Rep','Product'],values=['Total Revenue'],aggfunc=np.sum)
print(sales_data_pivot)
```

```python
test_pivot = pd.pivot_table(sales_data,index=["Rep"],columns=['Product'],values='Quantity')
print(test_pivot)
test_pivot.plot(kind="bar",stacked=True)
plt.show()
```

```python
sales_data_pivot = sales_data_pivot.unstack()
print(sales_data_pivot)
```
