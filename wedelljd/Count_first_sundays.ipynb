```python
import datetime
#define 

start_date = date(1901, 1, 1)
end_date = date(2000, 12, 31)

def get_first_sundays(start_date,end_date):
    counter=0
    if start_date.weekday()==6:
        counter+=1
    next_day = start_date+timedelta(days=1)
    while next_day <= end_date:
        #print(next_day.day,next_day.weekday())
        if next_day.day ==1 and next_day.weekday()==6:
            counter+=1
        next_day=next_day+timedelta(days=1)
    print(counter)
get_first_sundays(start_date,end_date)
```
