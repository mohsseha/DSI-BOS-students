## Regressions

* Perform a series of regressions on various combinations of the independent variables.
* Plot the predicted values versus the true values
* Which combinations gives the best fit?

```python
import statsmodels.api as sm
prediction_variables = ["sq__ft","baths"]
X=drop_sq_ft_zeroes[prediction_variables]
X=sm.add_constant(X)
y=drop_sq_ft_zeroes["price"]

model = sm.OLS(y, X).fit()
predictions = model.predict(X)
print(predictions)
model.summary()
```

```python
plt.scatter(predictions, y, s=30, c='r', marker='+', zorder=10)
plt.xlabel("Predicted Values from "+ str(prediction_variables))
plt.ylabel("Actual Values Price")
plt.show()
print "MSE:", model.mse_model
```
