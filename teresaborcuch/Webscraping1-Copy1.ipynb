```python
import requests
import bs4
from urllib import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
job_ids = []
descriptions = []
links = []
```

```python
URL = 'http://ontariohockeyleague.com/players/1'
soup = BeautifulSoup(urlopen(URL).read(), 'html.parser')
# <th class="table__th table__th-- text-col" title="Season" data-reactid=".0.0.0.3.0.0.1.2.0.0.0.$season_nameheader">Season</th><th class="table__th table__th-- text-col" title="Season" data-reactid=".0.0.0.3.0.0.1.2.0.0.0.$season_nameheader">Season</th>
print soup.prettify()
print soup.find_all('div', attrs={''})
```

```python



for i in np.arange(0,1000,10):
    URL = "https://www.indeed.com/jobs?q=data+scientist+$20,000&l=Boston&start={}&pp".format(str(i))
    #URL = "https://www.indeed.com/jobs?q=data+scientist+$20,000&l=Boston&start=10&pp"
    soup = BeautifulSoup(urlopen(URL).read(), 'html.parser')
    results = soup.find_all('div', attrs={'data-tn-component': 'organicJob'})
    for x in results:
        # get job_id
        job_id = x.find('h2', attrs={"class": "jobtitle"})['id']
        job_ids.append(job_id)
     
        # get job link
        link = "https://www.indeed.com" + x.find('h2', attrs={"class": "jobtitle"}).find('a')['href']
        links.append(link)    
            
        if 'rc/clk' in link:
            description = None
            descriptions.append(description)
        else:
            new_soup = BeautifulSoup(urlopen(link).read(), 'html.parser')
            description = new_soup.find('span', attrs = {'id': 'job_summary'}).text.strip()
            descriptions.append(description)   
```

```python
description_data = pd.DataFrame(columns = ['job_id','description','link'])
description_data['job_id'] = job_ids
description_data['description'] = descriptions
description_data['link'] = links

description_data.head(100)
```
