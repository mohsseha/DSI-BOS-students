```python
#load dataframe
import numpy as np
import pandas as pd
billboard = pd.read_csv("/Users/teresaborcuch/DSI-course-materials/curriculum/04-lessons/week-02/2.4-lab/assets/datasets/billboard.csv")
billboard
```

```python
# Make a column and a list of the number of weeks each song spent in the top 100
# Report summary stats
from scipy import stats
num_rows = range(317)
weeks_in_top_hun = []
for num in num_rows:
    row = billboard.iloc[num][7:]
    not_null = pd.notnull(billboard.iloc[num][7:])
    week_list = billboard.iloc[num][7:][not_null]
    num_weeks = len(week_list)
    weeks_in_top_hun.append(num_weeks)

billboard['weeks_in_top_hun'] = weeks_in_top_hun
billboard.head()
print "Maximum: ", max(weeks_in_top_hun)
print "Minumum: ", min(weeks_in_top_hun)
print "Mean: ", np.mean(weeks_in_top_hun)
print "Median: ", np.median(weeks_in_top_hun)
print "Mode: ", scipy.stats.mode(np.array(weeks_in_top_hun))[0][0]
print "Standard Deviation: ", np.std(weeks_in_top_hun)
print "Variance: ", np.var(weeks_in_top_hun)
```

```python
# summarize time
# do a string split on the :
billboard['time'] = billboard['time'].str.split(':')
billboard['time']
```

```python
time = billboard['time'].tolist()
```

```python
# convert minutes and seconds values into integers, multiply minutes by 60, and add
seconds_list = []
for pair in time:
    minutes = int(pair[0])
    seconds = int(pair[1])
    total_sec = (minutes * 60)+ seconds
    seconds_list.append(total_sec)
seconds_list
```

```python
billboard['time_in_sec'] = seconds_list
```

```python
# Descriptive statistics on time
print "Mean time: ", billboard['time_in_sec'].mean()
print "Standard deviation: ", np.std(billboard['time_in_sec'])
print "Median time: ", billboard['time_in_sec'].median()
print "Range: ", billboard['time_in_sec'].max() - billboard['time_in_sec'].min()
print "Mode time: ", scipy.stats.mode(billboard['time_in_sec'])[0][0]
print "Variance: ", np.var(billboard['time_in_sec'])
```

```python
# which genre has the highest average ranking of its songs?
genre_dict = {}
for num in num_rows:
    row = billboard.iloc[num]
    genre = row[4]
    not_null = pd.notnull(row[7:-2])
    ratings_series = row[7:-2][not_null]
    if genre in genre_dict:
        genre_dict[genre] = genre_dict[genre].append(ratings_series)
    if genre not in genre_dict:
        genre_dict[genre] = ratings_series
```

```python
genre_averages = {}
for item in genre_dict:
    genre_series = pd.Series(genre_dict.get(item))
    genre_mean = genre_series.mean()
    genre_averages[item] = genre_mean
    
genre_averages
# Average ranking the songs in each genre
```
