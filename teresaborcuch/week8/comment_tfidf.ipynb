```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# function to return an array of vectorized counts and feature names
# takes comments as df['comments'], i.e. a column in a dataframe

def get_tfidf(comments):
    tfidf = TfidfVectorizer(strip_accents = 'ascii', analyzer = 'word', stop_words = 'english', ngram_range = (1,1))
    values = tfidf.fit_transform(comments).todense()
    features = tfidf.get_feature_names()
    return values, features
    
```

```python
values, features = get_tfidf(init_data['comment'])
```
