```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv"

data = pd.read_csv(url)
print(data.columns)
```

```python
pd.get_dummies(data[' market ']).head()
```

```python
print(data[' funding_total_usd '].head(5))
```

```python
cleaned_funds_total = data[data[' funding_total_usd '] != '-'][' funding_total_usd '].apply(str).apply(str.strip).str.replace(",", "")
data['cleaned_funds_total'] = pd.to_numeric(cleaned_funds_total, errors = 'coerce')
data['cleaned_funds_total'].head(10)
```

```python
pd.pivot_table(data[data['cleaned_funds_total'].notnull()], values = ['cleaned_funds_total'], index = [' market '])
```
