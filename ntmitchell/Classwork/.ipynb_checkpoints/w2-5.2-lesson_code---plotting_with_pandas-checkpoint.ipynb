```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "../..https://github.com/mohsseha/DSI-course-materials/tree/master/curriculum/04-lessons/week-02/5.1-lesson/week-01/3.2-lesson/assets/datasets/sales.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
table_rep_price = pd.pivot_table(data[data['Status'] != 'declined'], values = ['Price'], index = ['Rep'], columns = ['Product'])
table_rep_price.plot.bar(stacked = True, figsize = (10,6))
plt.title("Sales by Rep, excluding declined sales")
plt.show()
```

```python
table_rep_price = pd.pivot_table(data, values = ['Quantity'], index = ['Rep'], columns = ['Product'])
table_rep_price.plot.bar(stacked = True, figsize = (10,6))
plt.title("Quantity sold by Rep, excluding declined sales")
plt.show()
```
