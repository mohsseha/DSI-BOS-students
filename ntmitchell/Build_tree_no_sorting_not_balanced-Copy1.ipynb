```python
def tree_build(elem, tree):
    if "node" not in tree:
        tree["node"] = elem
    else:
        if elem < tree["node"]:
            if "left" in tree:
                tree_build(elem, tree["left"])
            else:
                tree["left"] = {}
                tree["left"]["node"] = elem
        if elem >tree["node"]:
            if "right" in tree:
                tree_build(elem, tree["right"])
            else:
                tree["right"] = {}
                tree["right"]["node"] = elem
                
    return(tree)
```

```python
# Build list and RANDOMIZE order
import random
l = list(range(1000))
random.shuffle(l)

# Build tree
tree = {}
for elem in l:
    tree = tree_build(elem, tree)
```

```python
import time

runtimes = []

for i in range(1, 100):
    start = time.time()
    l = list(range(1000 * i))
    random.shuffle(l)

    # Build tree
    tree = {}
    for elem in l:
        tree = tree_build(elem, tree)
    end = time.time()
    runtimes.append(end - start)
    
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(runtimes)
plt.show()
```

```python
# Depth-first search
def finditem(tree, entry):
    flag = False
    
    if entry == tree["node"]:
        flag = True
    
    if entry < tree["node"]:
        if "left" in tree:
            flag = finditem(tree["left"], entry)
    if entry > tree["node"]:
        if "right" in tree:
            flag = finditem(tree["right"], entry)
     
    return(flag)
```
