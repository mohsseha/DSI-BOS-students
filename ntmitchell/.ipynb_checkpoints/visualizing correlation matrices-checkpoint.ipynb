```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
dataset = load_breast_cancer()
data_df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
filtered_data = data_df
for column in data_df.columns:
    if 'mean' in column:
        del filtered_data[column]
```

```python
filtered_data.head(10)
```

```python
filtered_data_correlations = filtered_data.corr()
filtered_data_correlations
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(filtered_data_correlations)
sns.plt.show()
```

```python
sns.pairplot(filtered_data)
sns.plt.show()
```

```python
import scipy.cluster.hierarchy as sch

d = sch.distance.pdist(filtered_data_correlations)

Z= sch.linkage(d,method='average')

P =sch.dendrogram(Z)
```
