```python
# Countries under primary classifications
ICO_composite_indicator_classes = {"Colombian Milds": """Colombia, Kenya, Tanzania""",
"Other Milds": """Bolivia, Burundi, Costa Rica, Cuba, Dominican Republic, Ecuador, El Salvador, Guatemala, Haiti, Honduras, India, Jamaica, Malawi, Mexico, Nepal, Nicaragua, Panama, Papua New Guinea, Peru, Rwanda, Venezuela, Zambia, Zimbabwe""",
"Brazilian Naturals": """Brazil, Ethiopia, Paraguay, Timor-Leste, Yemen""",
"Robustas": """Angola, Benin, Cameroon, Central African Republic, Dem. Rep. of Congo, Congo Rep., Côte d’Ivoire, Equatorial Guinea, Gabon, Ghana, Guinea, Guyana, Indonesia, Lao, People’s Dem. Rep., Liberia, Madagascar, Nigeria, Philippines, Sierra Leone, Sri Lanka, Thailand, Togo, Trinidad and Tobago, Uganda, Vietnam"""}

ICO_composite_indicator_classes_dataframe = pd.DataFrame(ICO_composite_indicator_classes, index = ["Country"]).T.Country.str.split(", ", expand = True).apply(lambda x: x.astype("category"))
ICO_composite_indicator_classes_dataframe = pd.get_dummies(ICO_composite_indicator_classes_dataframe, prefix="", prefix_sep="").T
```

For each country, record arabica and robusta production. 

```python
second_groupings_dataframe = pd.get_dummies(pd.DataFrame(second_groupings_dictionary).T, prefix = "", prefix_sep = "")
```

```python
for row in second_groupings_dataframe.index:
    ICO_composite_indicator_classes_dataframe.loc[row, :] = ICO_composite_indicator_classes_dataframe.loc[row, :] + second_groupings_dataframe.ix[row, 4:]

ICO_composite_indicator_classes_dataframe = ICO_composite_indicator_classes_dataframe.fillna(0).sort_index()

for feature in ICO_composite_indicator_classes_dataframe:
    ICO_composite_indicator_classes_dataframe.loc[:, feature] = ICO_composite_indicator_classes_dataframe.loc[:, feature].map({1: True, 0: False})
```
