### Import data

```python
production = Production_Data()
brazilian_naturals_ending_stock = production.get_ending_stocks_by_ICO_category()
```

```python
sns.boxplot(data = brazilian_naturals_ending_stock, x = "Country", y = "Percent Change")
plt.title("Distribution of Ending Stock Year-over-Year Percent Change")
plt.show()

pd.pivot_table(brazilian_naturals_ending_stock, values = ["Value (60kg bags)"], columns = ["Country"], index = ["Market_Year"]).plot(logy = True)
plt.title("Ending Stocks")
plt.show()

pd.pivot_table(brazilian_naturals_ending_stock, values = ["Percent Change"], columns = ["Country"], index = ["Market_Year"]).plot()
plt.title("Changes in Ending Stocks")
plt.show()
```
